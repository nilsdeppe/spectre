```python
def lagrange (x, i, xm):
    """
    Evaluates the i-th Lagrange polynomial at x
    based on grid data xm
    """
    y=1
    for j in range(len(xm)):
        if i != j:
            y *= (x - xm[j]) / (xm[i] - xm[j])
    return y
```
